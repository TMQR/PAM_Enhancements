```python
swarm_total = data.total
swarm_total = swarm_total.to_frame()
# data.plot()

swarm_total.index = pd.to_datetime(swarm_total.index)
swarm_total.info()
swarm_total.columns = ['swarm_total']
```

```python

swarm_total = swarm_total.add(200000)# Set margin estimate

swarm_total.rolling(50).min().plot()
```

```python
s = bt.Strategy('swarm_total', [bt.algos.RunQuarterly(),
                       bt.algos.SelectAll(),
#                        bt.algos.WeighInvVol(),
                       bt.algos.WeighEqually(),
                       bt.algos.Rebalance()])
```

```python
# create a backtest and run it
test = bt.Backtest(s, swarm_total)
```

```python
res = bt.run(test)
```

```python
res.plot(figsize=(10,7))
```

```python
res.display_monthly_returns()
```
